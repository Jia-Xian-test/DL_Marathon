## 1. 將 RGB 轉換為 HSV，分別印出 Hue, Saturation, Value

```python
# 將 RGB 轉換為 HSV
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt

img = cv2.imread('lena.png')
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv[:, :, [2, 1, 0]])
plt.axis('off')
plt.title('Origin')
plt.show()

# 分別印出 Hue, Saturation, Value
title = ['Hue', 'Saturation', 'Value']
plt.imshow(img_hsv[:, :, 0])
plt.axis('off')
plt.title('Hue')
plt.show()
plt.imshow(img_hsv[:, :, 1])
plt.axis('off')
plt.title('Saturation')
plt.show()
plt.imshow(img_hsv[:, :, 2])
plt.axis('off')
plt.title('Value')
plt.show()
```

## 2. 對 Value 進行 Histogram Equalization，並印出處理結果(處理完的圖片和處理完的 Value)

```python
# 先印出 HE 前的圖片與 value
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
img = cv2.imread('lena.png')
plt.imshow(img[:, :, [2, 1, 0]])
plt.axis('off')
plt.title('Origin')
plt.show()

# 印出 HE 後的圖片與 value
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
R, G, B = cv2.split(img)
output1_R = cv2.equalizeHist(R)
output1_G = cv2.equalizeHist(G)
output1_B = cv2.equalizeHist(B)
equ = cv2.merge((output1_R, output1_G, output1_B))
plt.imshow(equ[:, :, [2, 1, 0]])
plt.axis('off')
plt.title('After')
plt.show()
```

## 3. 將 RGB 轉換為 Gray level，再進行單值二值化及自適應二值化

```python
# 將圖片轉換為灰階圖片
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
img = cv2.imread('lena.png')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.title('Gray')
plt.show()
# Single Binarization
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
ret,img_bin=cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY)
plt.imshow(img_bin,cmap='gray')
plt.axis('off')
plt.title('Single Binarization')
plt.show()
# Adaptive Binarization
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
ret,img_adbin=cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(img_adbin,cmap='gray')
plt.axis('off')
plt.title('Adaptive Binarization')
plt.show()
```
