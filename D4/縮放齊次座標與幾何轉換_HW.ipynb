* 觀察圖片的變化

```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
img_path = 'lena.png'
img = cv2.imread(img_path)
plt.imshow(img[:,:,[2,1,0]])
plt.axis('off')
plt.title('Origin')
plt.show()
```

# 進行放大的比較

```python
# 方便顯示與進行縮放
interpolation_lst = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]
interpolation_name = ['INTER_LANCZOS4', 'INTER_LINEAR', 'INTER_AREA', 'INTER_CUBIC', 'INTER_LANCZOS4']

# 定義欲放大與縮小的尺寸
small_length, big_length = int(img.shape[0] / 4), int(4 * img.shape[0])

# 定義作圖版面大小
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的放大比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    """ Your Code Here: 插值演算法實作(Hint: cv2.resize) """
    interpolation=cv2.resize(img, (big_length, big_length),interpolation_lst[i])
    end = time.time() # 結束計時
    
    # 開始做圖
    plt.subplot(1, 5, i+1)
    plt.imshow(interpolation[:,:,[2,1,0]])
    plt.axis('off')
    plt.title('{} end in {:.4f}s'.format(interpolation_name[i], end - start))
plt.show()
```

# 進行縮小的比較

```python
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的縮小比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    """ Your Code Here: 插值演算法實作(Hint: cv2.resize) """
    interpolation=cv2.resize(img, (small_length, small_length),interpolation_lst[i])
    end = time.time() # 結束計時
    
    # 開始做圖
    plt.subplot(1, 5, i+1)
    plt.imshow(interpolation[:,:,[2,1,0]])
    plt.axis('off')
    plt.title('{} end in {:.4f}s'.format(interpolation_name[i], end - start))
plt.show()
```
